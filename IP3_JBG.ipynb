```python
#4
Car_Sales=pd.read_csv("car-sales.csv")
```

```python
Car_Sales.head()
```

```python
Car_Sales.columns
```

```python
#4
Car_Sales['Price'].plot()
```

```python
Car_Sales['Odometer (KM)'].plot()
```

```python
# One works because it is a proper integer, while the price is not a valid integer due to the dollar sign. 
```

```python
#5
Car_Sales['Price']=Car_Sales['Price'].str.replace('[\$,]','',regex=True).astype('int64')
#Remove dollar sign and convert to 64 bit integer
```

```python
Car_Sales.dtypes
```

```python
#6
Car_Sales['Total Sales']=Car_Sales['Price'].cumsum()
Car_Sales
```

```python
#7
from datetime import datetime
date_range=pd.date_range(start=datetime.now(),periods=len(Car_Sales))
Car_Sales['Sale Date']=date_range
Car_Sales
```

```python
#8
plt.plot(Car_Sales['Price'],Car_Sales['Total Sales'])
plt.xlabel('Price')
plt.ylabel('Total Sales')
plt.title('Total Sales vs Price Line Plot')
```

```python
#9
plt.scatter(Car_Sales['Odometer (KM)'],Car_Sales['Price'])
plt.xlabel('Miles Driven (KM)')
plt.ylabel('Price')
plt.title("Odometer reading in Kilometers vs Price")
```

```python
#10
plt.bar(Car_Sales['Make'],Car_Sales['Odometer (KM)'])
plt.xlabel('Make/Brand')
plt.ylabel('Odometer Reading in Kilometers')
plt.title("Odometer Reading in Kilometers in Relation to Make/Brand")
```

```python
#11
plt.hist(Car_Sales['Make'])
plt.xlabel('Make/Model')
plt.ylabel('Frequency')
plt.title('Car Make/Model and Sales')
```

```python
#12
fig=plt.figure(figsize=(16,16))
ax1=fig.add_subplot(221)
ax1.plot(Car_Sales['Price'],Car_Sales['Total Sales'])
ax2=fig.add_subplot(222)
ax2.scatter(Car_Sales['Odometer (KM)'],Car_Sales['Price'])
ax3=fig.add_subplot(223)
ax3.bar(Car_Sales['Make'],Car_Sales['Odometer (KM)'])
ax4=fig.add_subplot(224)
ax4.hist(Car_Sales['Make'])
```

```python
#13
sns.boxplot(x='Colour',y='Total Sales',data=Car_Sales)
```

```python
#14
sns.lmplot(x='Total Sales',
           y='Odometer (KM)',
          data=Car_Sales,
          ci=False,
          hue='Make')
```

```python
#15
sns.pairplot(Car_Sales)
```

```python
#16
heat_map=Car_Sales[['Total Sales','Price','Odometer (KM)']].corr(method='pearson')
cols=['Total Sales','Price','Odometer (KM)']
sns.heatmap(heat_map, xticklabels=cols,yticklabels=cols,annot=True)
```

 There is a negative correlation with Total sales and odometer, a very negative correlation with odometer and price, and a slightly positive correlation with total sales and price.
